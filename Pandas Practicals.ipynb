### Merging the Data

```python
market = pd.read_csv('market_fact.csv')
customer = pd.read_csv('cust_dimen.csv')
product = pd.read_csv('prod_dimen.csv')
```

```python
market.head()
```

```python
customer.head()
```

```python
product.head()
```

```python
#market - 'Cust_id'
#customer - 'Id'
```

```python
#pd.merge(market, customer, left_on = 'Cust_id', right_on = 'Id', how = 'inner')
```

```python
new_df = pd.merge(market, customer, on = 'Cust_id', how = 'inner')
new_df.head()
```

```python
final_df = pd.merge(new_df, product, on = 'Prod_id', how = 'inner')
final_df.head()
```

```python
final_df.to_csv('merged_data.csv', index = False)
```

### Concatenating the Data

It is a more straightforward and simple process where we combine the data.

```python
df1 = pd.DataFrame({'Cust_id': [101,102,103,104,105],
                   'Prod_id': ['P1', 'P2', 'P3', 'P4', 'P5'],
                   'Sales': [100, 134, 145, 230, 43]})
```

```python
df2 = pd.DataFrame({'Cust_id': [106,107,108,109,110],
                   'Prod_id': ['P6', 'P7', 'P8', 'P9', 'P10'],
                   'Sales': [10, 138, 165, 220, 43]})
```

```python
df1
```

```python
df2
```

```python
concat_data = pd.concat([df1, df2], axis = 0, ignore_index = True)
```

```python
concat_data
```

```python
#remove the duplicates
concat_data.drop_duplicates()
```

```python
#delete a specific row using its index number
concat_data.drop([10])
```

```python
df3 = pd.DataFrame({'Profit': [10, 12, 14, 17, 18], 
                   'Discount': [0.1, 0.2, 0.01, 0.04, 0.3]})
```

```python
df3
```

```python
df1
```

```python
pd.concat([df1, df3], axis = 1)
```
